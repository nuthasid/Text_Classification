```python
import matplotlib.pyplot as plt
import numpy
import csv
import os
import pandas as pd
from scipy import stats
```

```python
data_path = '..\\nectec\\concat_doc_length.txt'
with open(data_path, 'r') as dp:
    csv_file = csv.reader(dp, delimiter='\t')
    doc_length = [int(line[0]) for line in csv_file]
```

```python
print(doc_length)
```

```python
data = numpy.asarray(doc_length)
print(data)
```

```python
pd.DataFrame(data).plot(kind='density', xlim=(1,500))
```

```python
pd.DataFrame(data).plot(kind='hist', xlim=(1,500), bins=500)
```
