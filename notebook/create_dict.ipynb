```python
## clean text
file_name = 'masterDB_JPA Data - 20180406_flatten.json'
new_name = 'data20180620_cleaned.json'
dict_file = 'dict_2gram.json'
field = 'desc'
processes = 8

import sys
if not '..' in sys.path:
    sys.path.append('..')
import os
import json
import time
from create_dict import CreateDict

start = time.time()

Create_Dict = CreateDict(n_gram=2, processes=processes, stop_en='en_stop_word.txt', stop_th='th_stop_word.txt')

data_feed = []
with open(os.path.join(os.getcwd(), '..', 'data', file_name), 'rt', encoding='utf-8') as f:
    for line in f.readlines():
        job_ad = json.loads(line)
        data_feed.append(job_ad[field])
        
dicts = Create_Dict.create_dict(data_feed)

tokens = {}
for doc in dicts:
    for token in doc:
        if tokens.get(token):
            tokens[token] += 1
        else:
            tokens[token] = 1
print(len(tokens))
with open(os.path.join(os.getcwd(), '..', 'data', dict_file), 'w', encoding='utf-8') as f:
    json.dump(tokens, f, ensure_ascii=False)
```
