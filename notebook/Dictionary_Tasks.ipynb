```python
## Load Pickle to python object
## pickle object has dictionary from data 
## masterDB = masterDB_JPA Data - 20180406_flatten.json

import pickle
import os

cwd = os.getcwd()

title_dict = pickle.load(open(cwd+"/../data/"+"title.pck", 'rb'))
desc_dict = pickle.load(open(cwd+"/../data/"+"desc.pck",'rb'))

## create dictionary list -> list and sort

def getVal(mytuple):
    return mytuple[1]

title_dict = list(title_dict.items())
desc_dict = list(desc_dict.items())
title_dict = sorted(title_dict, key=getVal, reverse=True)
desc_dict = sorted(desc_dict, key=getVal, reverse=True)
```

```python
## improve dictionary

## remove signs
def getList(range_in):
    return [item for item in range_in]

def isSign(word, threshold=1):
    count = 0
    sign = getList(range(33,47+1))+getList(range(48,57+1))+getList(range(58,64+1))+getList(range(91,96+1))
    for char in word:
        if(ord(char) in sign):
            count+=1
            if(count == threshold or len(word) == 1):
                return True
    else:
        return False
    
title_dict_sign = [item for item in title_dict if isSign(item[0],1)]
desc_dict_sign = [item for item in desc_dict if isSign(item[0],1)]

title_newdict = [item for item in title_dict if not isSign(item[0],1)]
desc_newdict = [item for item in desc_dict if not isSign(item[0],1)]
```

```python
def reduceByValue(inlist):
  mydict = dict()
  for item in inlist:
    if(mydict.get(item[1])):
      mydict[item[1]]+=1
    else:
      mydict[item[1]]=1
  return mydict
```
